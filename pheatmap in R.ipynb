```r
library(pheatmap)
```

```r
geneseries <- read.csv('toR_selected.tsv', sep = '\t', header=T, row.names= 1)
```

```r
pheatmap(geneseries, cluster_rows = F, cluster_cols = F, filename = "test.tiff", show_rownames = F)
```

```r
pheatmap(geneseries, cluster_rows = F, cluster_cols = F, show_rownames = F)
```

```r
svg(file="heatmap_all.svg")
pheatmap(geneseries, cluster_rows = F, cluster_cols = F)
dev.off()
```

* **Clustered image has been in folder, called 'test'.**
